```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Load the original color image (BGR format by default)
img = cv2.imread('images/input.jpg')

levels = 4  # Example: reduce to 4 intensity levels
factor = 256 // levels

# Apply reduction to each channel separately
reduced = (img // factor) * factor

# Create folder and save result
os.makedirs('outputs/task1', exist_ok=True)
cv2.imwrite(f'outputs/task1/reduced_{levels}_levels.jpg', reduced)

# Display using RGB format
plt.imshow(cv2.cvtColor(reduced, cv2.COLOR_BGR2RGB))
plt.title(f"Reduced to {levels} Intensity Levels (Color)")
plt.axis('off')
plt.show()
```
