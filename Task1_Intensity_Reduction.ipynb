```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Load the original color image (BGR format)
img = cv2.imread('images/input.jpg')

levels = 4  # Example: reduce to 4 intensity levels
factor = 256 // levels

# Apply reduction to each channel
reduced = (img // factor) * factor

# Create output directory
os.makedirs('outputs/task1', exist_ok=True)
cv2.imwrite(f'outputs/task1/reduced_{levels}_levels.jpg', reduced)

# Display both images in one row
plt.figure(figsize=(10, 5))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

# Reduced intensity image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(reduced, cv2.COLOR_BGR2RGB))
plt.title(f"Reduced to {levels} Levels")
plt.axis('off')

plt.tight_layout()
plt.show()
```
