```python
import cv2, numpy as np
import matplotlib.pyplot as plt
import os

img = cv2.imread('images/input.jpg', cv2.IMREAD_GRAYSCALE)

levels = 4  # example
factor = 256 // levels
reduced = (img // factor) * factor

#  for Save output
os.makedirs('outputs/task1', exist_ok=True)
cv2.imwrite(f'outputs/task1/reduced_{levels}_levels.jpg', reduced)

plt.imshow(reduced, cmap='gray')
plt.title(f"Reduced to {levels} levels")
plt.axis('off')
plt.show()
```
