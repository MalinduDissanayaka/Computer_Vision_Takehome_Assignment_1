```python
import cv2, numpy as np
import matplotlib.pyplot as plt
import os

kernel_sizes = [3, 10, 20]

img = cv2.imread('images/input.jpg', cv2.IMREAD_GRAYSCALE)
os.makedirs('outputs/task2', exist_ok=True)

for k in kernel_sizes:
    blurred = cv2.blur(img, (k, k))
    cv2.imwrite(f'outputs/task2/blur_{k}x{k}.jpg', blurred)
    plt.imshow(blurred, cmap='gray')
    plt.title(f"{k}x{k} Average Blur")
    plt.axis('off')
    plt.show()
```
