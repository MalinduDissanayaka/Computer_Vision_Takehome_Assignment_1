```python
import cv2, numpy as np
import matplotlib.pyplot as plt
import os

def block_average(img, block_size):
    h, w = img.shape
    output = img.copy()
    for i in range(0, h - block_size + 1, block_size):
        for j in range(0, w - block_size + 1, block_size):
            block = img[i:i+block_size, j:j+block_size]
            avg = int(np.mean(block))
            output[i:i+block_size, j:j+block_size] = avg
    return output

img = cv2.imread('images/input.jpg', cv2.IMREAD_GRAYSCALE)
os.makedirs('outputs/task4', exist_ok=True)

for b in [3, 5, 7]:
    out = block_average(img, b)
    cv2.imwrite(f'outputs/task4/block_avg_{b}x{b}.jpg', out)
    plt.imshow(out, cmap='gray')
    plt.title(f"{b}x{b} Block Averaging")
    plt.axis('off')
    plt.show()
```
